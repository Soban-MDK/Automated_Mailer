```python
from google.oauth2 import service_account
from googleapiclient.discovery import build

# Path to your credentials.json file
SERVICE_ACCOUNT_FILE = 'credentials.json'

# Define the scope for Google Sheets API
SCOPES = ['https://www.googleapis.com/auth/spreadsheets']

# Create credentials using the service account file
credentials = service_account.Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES)

# Build the Sheets API service
service = build('sheets', 'v4', credentials=credentials)

# The ID of your spreadsheet (found in the URL of your Google Sheet)
SPREADSHEET_ID = '1G9UHL_grR0RaLwVQK1Gy36JPd5iHHK7sBncNYXZb9S4'

# Example: Reading data from a range in the spreadsheet
range_name = 'Sheet1!A1:C37'
result = service.spreadsheets().values().get(
    spreadsheetId=SPREADSHEET_ID, range=range_name).execute()

values = result.get('values', [])

# Converting to a DataFrame
df = pd.DataFrame(values[1:], columns=values[0])

# Display the DataFrame
df
```

## USING GSPREAD

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# Define the scope and path to the credentials
scope = ['https://spreadsheets.google.com/feeds',
         'https://www.googleapis.com/auth/drive']

creds = ServiceAccountCredentials.from_json_keyfile_name('credentials.json', scope)

# Authorize the client
client = gspread.authorize(creds)

# Open a spreadsheet by title or by key (spreadsheet ID)
spreadsheet = client.open_by_key('1G9UHL_grR0RaLwVQK1Gy36JPd5iHHK7sBncNYXZb9S4')

# Select the first worksheet
worksheet = spreadsheet.get_worksheet(0)

# Get all records
records = worksheet.get_all_records()

# Convert the records to a DataFrame
df = pd.DataFrame(records)

df
```
